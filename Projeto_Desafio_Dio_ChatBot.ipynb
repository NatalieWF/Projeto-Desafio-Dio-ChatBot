```python
import os
from pypdf import PdfReader
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Carregar os PDFs e extrair o texto
def load_and_extract_text(pdf_paths):
    text_list = []
    for path in pdf_paths:
        try:
            reader = PdfReader(path)
            text = ""
            for page in reader.pages:
                text += page.extract_text() + "\n"
            text_list.append(text)
        except Exception as e:
            print(f"Erro ao processar o arquivo {path}: {e}")
    return text_list
```

```python

# Simule o upload de arquivos PDF no Colab
# Crie alguns arquivos PDF de exemplo (você pode substituir pelos seus)
with open("arquivo1.pdf", "w") as f:
    f.write("Este é o conteúdo do primeiro arquivo PDF. Fala sobre IA e aprendizado de máquina.")

with open("arquivo2.pdf", "w") as f:
    f.write("O segundo PDF discute processamento de linguagem natural e chatbots.")

pdf_files = ["arquivo1.pdf", "arquivo2.pdf"]
document_texts = load_and_extract_text(pdf_files)

# Se você for usar arquivos do Google Drive, descomente e execute este bloco:
# from google.colab import drive
# drive.mount('/content/drive')
# pdf_files = ["/content/drive/MyDrive/seu_arquivo1.pdf", "/content/drive/MyDrive/pasta/seu_arquivo2.pdf"]
# document_texts = load_and_extract_text(pdf_files)
```

```python
# 2. Criar embeddings para os textos dos documentos
model = SentenceTransformer('all-MiniLM-L6-v2')
document_embeddings = model.encode(document_texts)
```

```python
# 3. Função para buscar respostas relevantes
def buscar_resposta(query, document_texts, document_embeddings, top_n=3):
    query_embedding = model.encode([query])
    similarities = cosine_similarity(query_embedding, document_embeddings)[0]

    # Combine textos e similaridades para ordenação
    doc_similarity_pairs = list(zip(document_texts, similarities))
    # Ordene por similaridade em ordem decrescente
    sorted_pairs = sorted(doc_similarity_pairs, key=lambda x: x[1], reverse=True)

    relevant_texts = [pair[0] for pair in sorted_pairs[:top_n]]
    return relevant_texts
```

```python
# 4. Criar um chatbot interativo (simples)
def chatbot():
    print("Olá! Faça sua pergunta sobre os documentos.")
    while True:
        query = input("> ")
        if query.lower() in ["sair", "fim", "adeus"]:
            print("Até logo!")
            break

        relevant_chunks = buscar_resposta(query, document_texts, document_embeddings)

        if relevant_chunks:
            print("Textos relevantes encontrados:")
            for i, chunk in enumerate(relevant_chunks):
                print(f"[{i+1}] \"{chunk[:100]}...\"") # Exibe os primeiros 100 caracteres

            # Aqui, em um cenário mais avançado, você usaria um modelo de linguagem maior
            # (como o Gemini Pro do Google AI Studio ou o GPT-3.5/4 da OpenAI)
            # para gerar uma resposta concisa baseada nesses textos relevantes.
            # Para simplificar, vamos apenas exibir os textos relevantes.
            print("\n(Implementação futura: Uma resposta concisa seria gerada aqui com base nos textos relevantes.)")
        else:
            print("Nenhum texto relevante encontrado para sua pergunta.")

# Iniciar o chatbot
chatbot()
```
